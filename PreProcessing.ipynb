```python
import pandas as pd
import numpy as np
import matplotlib as plt
df = pd.read_csv("hack_ru.csv")
df
df["total_pack_size_ml_g"] = df["total_pack_size_ml_g"]/df["unit_pack_size_ml_g"]
```

```python
allIngred = []
for x in df["ingredients"]:
    try:
        allIngred.extend([y.strip(" ") for y in x.split(",")])
    except:
        print (x)

print (len(allIngred))
newIng = []
for x in allIngred:
    new= x.strip("(").strip(")").lower()
    if ("ingredient" not in new):
        newIng.append(new)
        
newIng = list(set(newIng))
len(newIng)
allIngred = newIng.copy()
for i, x in enumerate(newIng):
    newIng[i] = newIng[i].split("(")[0].strip(" ")
newIng = list(set(newIng))
len(newIng)
for i, x in enumerate(newIng):
    
    newIng[i] = newIng[i].split(":")[0].strip(" ")
newIng = list(set(newIng))
len(newIng)
for i, x in enumerate(newIng):
    newIng[i] = newIng[i].split("*")[0].strip(" ")
newIng = list(set(newIng))
for i, x in enumerate(newIng):
    newIng[i] = newIng[i].split('"')[0].strip(" ")
newIng = list(set(newIng))
for x in (newIng.copy()):
    if len(x) < 4:
        newIng.remove(x)
len(newIng)
```

```python
def spotOf(ing, ings):
    for i,x in enumerate(ings.split(",")):
        if(ing in x.lower()):
            return i
    return -1
for x in newIng:
    print (x)
    if (x not in df.columns):
        df[x] = df.fillna("n/a").apply(lambda row: spotOf(x, row["ingredients"]), axis = 1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("toothpaste_full.csv", error_bad_lines=False, sep = "\t")
```

```python
dummy = pd.get_dummies(df.country)
new_cols = []
for x in dummy.columns:
    new_cols.append("country_"+ x)
dummy.columns = new_cols
dummy

dummy2 = pd.get_dummies(df.company)
final = pd.concat([df,dummy,dummy2], axis =1)
cols = list(final.columns)
cols.remove("country")
cols.remove("company")
cols.remove("Unnamed: 0")
cols.remove("ingredients")

final = final[cols]
final = final.fillna(1)
train_cols = list(final.columns)
test_cols = ["price_per_100g_ml_dollars"]
train_cols.remove("price_per_100g_ml_dollars")


train_features, test_features, train_labels, test_labels =train_test_split(final[train_cols], final[test_cols], test_size = 0.3, random_state = 42)
train_features
```

```python
from sklearn.svm import SVR
clf = SVR(kernel="linear", C=1.0, epsilon=0.2)
clf.fit(train_features, train_labels)  
```
